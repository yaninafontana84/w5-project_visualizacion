```python
platform_mapping = {
    'Wii': 'Nintendo Wii',
    'NES' : 'NES (Nintendo Entertainment System)', 
    'GB' : 'Nintendo Game Boy', 
    'DS':'Nintendo Dual Screen', 
    'X360': 'Xbox360',
    'PS3': 'PlayStation3', 
    'PS2': 'PlayStation2', 
    'SNES': 'Super Nintendo Entertainment System', 
    'GBA' : 'Game Boy Advance',
    'PS4': 'PlayStation4', 
    '3DS': 'Nintendo 3DS', 
    'N64': 'Nintendo 64', 
    'PS': 'PlayStation', 
    'XB': 'Xbox', 
    '2600': 'Atari 2600', 
    'PSP': 'PlayStation Portable', 
    'XOne': 'Xbox One',
    'WiiU' : 'Nintendo WiiU', 
    'GC' : 'Nintendo GameCube', 
    'GEN': 'Sega Genesis', 
    'DC': 'Sega Dreamcast', 
    'PSV': 'PlayStation Vita', 
    'SAT': 'Sega Saturn', 
    'SCD': 'Sega CD', 
    'WS': 'WonderSwan', 
    'NG': 'Neo Geo', 
    'TG16': 'TurboGrafx-16', 
    'GG': 'Game Gear', 
    'PCFX': 'NEC PC-FX'
}

video_games_sales['Platform'] = video_games_sales['Platform'].replace(platform_mapping)
```

```python
column_mapping = {
    'Name': 'Nombre',
    'Platform':'Plataforma',
    'Year_of_Release': 'A√±o',
    'Genre': 'G√©nero',
    'Publisher': 'Compa√±√≠a',
    'NA_Sales': 'USA',
    'EU_Sales': 'Europa',
    'JP_Sales': 'Jap√≥n',
    'Other_Sales': 'Otros'
}

video_games_sales = video_games_sales.rename(columns=column_mapping)
```
