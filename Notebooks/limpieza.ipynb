```python
# librerias a importar

import pandas as pd
pd.set_option('display.max_columns', None)


import warnings
warnings.filterwarnings('ignore')
```

```python
video_games_sales_ori = pd.read_csv('../data/video_games_sales.csv', encoding='latin1') #leo el archivo "INE"

video_games_sales = video_games_sales_ori.copy() #hago una copia

video_games_sales.head() #miro qu√© pinta tiene para empezar a limpiar
```

```python
video_games_sales.columns
```

```python
columnas_a_eliminar = ['Global_Sales', 'Critic_Score',
       'Critic_Count', 'User_Score', 'User_Count', 'Developer', 'Rating']

video_games_sales = video_games_sales.drop(columnas_a_eliminar, axis=1) 
```

```python
video_games_sales
```

```python
video_games_sales.info()
```

```python
video_games_sales_cleaned = video_games_sales.dropna()
```

```python
video_games_sales_cleaned.info()
```

```python
video_games_sales_cleaned
```

```python

video_games_sales_cleaned['Year_of_Release'] = video_games_sales_cleaned['Year_of_Release'].astype(int)
video_games_sales_cleaned
```

```python
video_games_sales_cleaned.to_csv('video_games.csv', index=False) 
```

```python
from sqlalchemy import create_engine
```

```python
with open ('pw.txt') as file:
    password = file.read()
```

```python
str_conn = f'mysql+pymysql://root:{password}@localhost:3306/video_games'
cursor = create_engine(str_conn)
```

```python
video_games_sales_cleaned.to_sql(name='video_games_sales',       # nombre de la tabla
            con=cursor,           # conexion al servidor
            if_exists='replace',  # replace sobreescribe la tabla
            index=True)
```
